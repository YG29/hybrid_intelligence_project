```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score
```

```python
CSV_PATH = "human_annotations.csv"
```

```python
df = pd.read_csv(CSV_PATH)
```

```python
df = df.copy()
df["y_true"] = df["true_label_index"]
df["y_pred"] = df["human_label_index"]
```

```python
acc = accuracy_score(df["y_true"], df["y_pred"])
f1  = f1_score(df["y_true"], df["y_pred"], average="macro")

print("=== Human Accuracy ===")
print(f"Samples: {len(df)}")
print(f"Accuracy: {acc:.6f}")
print(f"F1 (macro): {f1:.6f}")
```
